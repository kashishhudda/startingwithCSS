1. Create a simple page with some div tags and show different ways to add CSS as well as what happens when you target the same elements with inline, internal, and external CSS. Also, utilize comments in the project where required.

```python
< !DOCTYPE html>
< html lang="en">
< head>
    < meta charset="UTF-8">
    < meta name="viewport" content="width=device-width, initial-scale=1.0" >
    < title>CSS< /title>
    < link rel="stylesheet" href="style.css" >
    < style>
        .box{
            background-color: rgb(177, 177, 71);
            color: black;
            font-size: larger;
        }
    < /style>
< /head>
< body>
    < h1>CSS Demonstration: Inline VS External VS internal< /h1>
    <!-- Inline CSS overrides both internal and external CSS -->
    < div class="box" style="background-color: rgb(144, 103, 182);color: rgb(59, 7, 7);" >
        This is a div with < strong>inline CSS< /strong> applied.
    < /div>
    <!-- Internal CSS is applied here because inline is not used -->
    < div class="box">This div uses < strong>internal CSS< /strong>.
    < /div>
    <!-- Only external CSS applies here -->
    < div class="note">This div shows styles from < strong>external CSS< /strong>
    < /div>
< /body>
< /html>

"""External CSS"""

/* External CSS */
.box {
    background-color: rgb(238, 116, 136);   /* This will be overridden by internal or inline */
    color: rgb(104, 30, 30);
    border: 2px solid rgb(80, 47, 85);
    padding-top: 3px;
    margin-top: 3px;
}

.note {
    color: rgb(7, 10, 7);
    text-decoration: underline;
    margin-top:3px;
    font-size: larger;
}
```

2. Build an HTML page with multiple paragraphs, each assigned a unique class name. Write CSS rules using class selectors to apply distinct styling to each paragraph. Follow the BEM naming convention and explain how you've named the classes.

```python
"""HTML"""
< !DOCTYPE html>
< html lang="en">
< head>
    < meta charset="UTF-8">
    < meta name="viewport" content="width=device-width, initial-scale=1.0" >
    < title>BEM Naming Convention</title>
    < style>
        /* CSS using BEM naming convention */
        .paragraph {
            font-family: sans-serif;
            margin-bottom: 1em;
            padding: 0.5em;
            border: 1px solid #ccc;
        }

        .paragraph--primary {
            background-color: #e0f7fa; /* Light cyan */
            color: #004d40; /* Dark teal */
        }

        .paragraph--secondary {
            background-color: #fff3e0; /* Light orange */
            color: #e65100; /* Dark orange */
        }

        .paragraph--highlight {
            background-color: #fce4ec; /* Light pink */
            color: #ad1457; /* Dark pink */
            font-weight: bold;
        }
    < /style>
< /head>
< body>
    < h1>BEM Naming Convention Example< /h1>

    < p class="paragraph paragraph--primary">
        This is the first paragraph. It uses the < code>paragraph--primary< /code> class to apply a light cyan background and dark teal text color.
    < /p>

    < p class="paragraph paragraph--secondary">
        This is the second paragraph. It uses the < code>paragraph--secondary< /code> class for a light orange background and dark orange text color.
    < /p>

    < p class="paragraph paragraph--highlight">
        This is the third paragraph. It uses the < code>paragraph--highlight< /code> class for a light pink background, dark pink text, and bold font.
    < /p>

< /body>
< /html>
```

3. Develop an HTML form with various input elements. Use CSS to style the form, including setting background colors for input fields. Create a custom color palette for the form elements, and demonstrate how to apply opacity to one of the form sections.

```python
< !DOCTYPE html>
< html lang="en">
< head>
  < meta charset="UTF-8">
  < meta name="viewport" content="width=device-width, initial-scale=1.0">
  < title>Styled Form Example< /title>
  < style>
    /* Custom Color Palette */
    :root {
      --primary: #4A90E2;
      --secondary: #50E3C2;
      --accent: #F5A623;
      --light-bg: #f9f9f9;
      --input-bg: #ffffff;
      --input-border: #cccccc;
    }

    body {
      font-family: Arial, sans-serif;
      background-color: var(--light-bg);
      display: flex;
      justify-content: center;
      padding: 20px;
    }

    form {
      background-color: #fff;
      padding: 20px;
      border-radius: 12px;
      box-shadow: 0 4px 10px rgba(0,0,0,0.1);
      width: 400px;
    }

    h2 {
      text-align: center;
      color: var(--primary);
    }

    label {
      display: block;
      margin: 10px 0 5px;
      font-weight: bold;
      color: var(--accent);
    }

    input, select, textarea {
      width: 100%;
      padding: 10px;
      margin-bottom: 15px;
      border: 1px solid var(--input-border);
      border-radius: 6px;
      background-color: var(--input-bg);
      transition: border-color 0.3s;
    }

    input:focus, select:focus, textarea:focus {
      border-color: var(--primary);
      outline: none;
    }

    /*  Section with opacity */
    .semi-transparent-section {
      background-color: var(--secondary);
      padding: 15px;
      border-radius: 8px;
      opacity: 0.8; /*  Opacity applied */
      margin-bottom: 20px;
    }

    button {
      width: 100%;
      padding: 12px;
      background-color: var(--primary);
      color: #fff;
      font-size: 16px;
      font-weight: bold;
      border: none;
      border-radius: 8px;
      cursor: pointer;
      transition: background-color 0.3s;
    }

    button:hover {
      background-color: var(--accent);
    }
  < /style>
< /head>
< body>

  < form>
    < h2>Registration Form< /h2>

    < div class="semi-transparent-section">
      < label for="fullname">Full Name:< /label>
      < input type="text" id="fullname" name="fullname" placeholder="Enter your name">

      < label for="email">Email:< /label>
      < input type="email" id="email" name="email" placeholder="Enter your email">

      < label for="password">Password:< /label>
      < input type="password" id="password" name="password" placeholder="Enter your password">
    < /div>

    < label for="gender">Gender:< /label>
    < select id="gender" name="gender">
      < option value="">-- Select --< /option>
      < option value="male">Male< /option>
      < option value="female">Female< /option>
      < option value="other">Other< /option>
    < /select>

    < label for="bio">Short Bio:< /label>
    < textarea id="bio" name="bio" rows="4" placeholder="Tell us about yourself...">< /textarea>

    < label>Subscribe to Newsletter:< /label>
    < input type="checkbox" id="subscribe" name="subscribe"> Yes, keep me updated!

    < button type="submit">Submit< /button>
  < /form>

< /body>
< /html>
```
